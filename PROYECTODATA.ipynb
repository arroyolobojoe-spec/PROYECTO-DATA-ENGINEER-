```python
 nombre_producto = 'bicicleta de montaña AW-2026'
precio_base = 1200.50
cantidad_pedida= 3
tasa_impuesto= 0.19


# operaciones matematicas

subtotal = precio_base * cantidad_pedida
total_impuesto= subtotal * tasa_impuesto
precio_final= subtotal + total_impuesto


# Resultados:

print('--- DETALLE DE COMPRA---')
print( 'Producto:', nombre_producto)
print('El subtotal es: COP',subtotal, '$')
print ('Impuesto totales: COP', total_impuesto, '$')
print ('Precio Final: COP', precio_final, '$')




# Una lista que contiene diccionarios. Cada diccionario es un pedido de AdventureWorks.
pedidos_ingresados = [
    {"producto": "Bicicleta de Ruta", "precio": 1500.00, "cantidad": 2},
    {"producto": "Casco Premium", "precio": -50.00, "cantidad": 1},      # ERROR: Precio negativo
    {"producto": "Guantes de Ciclismo", "precio": 25.00, "cantidad": 0},  # ERROR: Cantidad cero
    {"producto": "Llantas Todo Terreno", "precio": 45.00, "cantidad": 4}
]

print("--- INICIANDO PROCESAMIENTO DE PIPELINE ---")

for pedido in pedidos_ingresados:
   if pedido['precio'] <= 0:

      print(f"❌ ALERTA: Pedido rechazado para '{pedido['producto']}'. Precio inválido: ${pedido['precio']}. Cantidad. {pedido['cantidad']}")

   elif pedido['cantidad'] <= 0:
        print(f"❌ ALERTA: Pedido rechazado para '{pedido['producto']}'. Cantidad inválida: {pedido['cantidad']}. Precio: ${pedido['precio']}")


   else:
        subtotal = pedido['precio'] * pedido['cantidad']
        print(f" PROCESADO: {pedido['producto']}. TOTAL: {subtotal}. Cantidad: {pedido['cantidad']}")


print("--- FIN DEL REPORTE ---")


```
