```python
 nombre_producto = 'bicicleta de montaña AW-2026'
precio_base = 1200.50
cantidad_pedida= 3
tasa_impuesto= 0.19


# operaciones matematicas

subtotal = precio_base * cantidad_pedida
total_impuesto= subtotal * tasa_impuesto
precio_final= subtotal + total_impuesto


# Resultados:

print('--- DETALLE DE COMPRA---')
print( 'Producto:', nombre_producto)
print('El subtotal es: COP',subtotal, '$')
print ('Impuesto totales: COP', total_impuesto, '$')
print ('Precio Final: COP', precio_final, '$')



```
